Bar Plot for cities with best median Income to Price Per Sqr foot

```python
import matplotlib.pyplot as plt
import seaborn as sns

housing_data = pd.read_csv('/American_Housing_Data_20231209.csv')
# Calculate price per living space (in dollars per square foot) and group data by city
housing_data['Price Per Living Space'] = housing_data['Price'] / housing_data['Living Space']
city_grouped_data = housing_data.groupby('City').agg({
    'Median Household Income': 'median',
    'Price Per Living Space': 'mean'
}).reset_index()

# Calculate the ratio of median household income to price per living space for each city
city_grouped_data['Income to Price Ratio'] = city_grouped_data['Median Household Income'] / city_grouped_data['Price Per Living Space']

# Sort the cities by their income to price ratio and select the top 8
top_cities = city_grouped_data.sort_values(by='Income to Price Ratio', ascending=False).head(8)

# Create a factorplot for these top 5 cities
plt.figure(figsize=(10, 6))
sns.catplot(x='City', y='Income to Price Ratio', kind='bar', data=top_cities, height=6, aspect=2)
plt.title('Top 5 Cities with Best Median Income to Price Per Living Space Ratios')
plt.ylabel('Income to Price Ratio')
plt.xlabel('City')
plt.xticks(rotation=45)
plt.show()
```

Comparing 2 cities

```python
# Filter the dataset for New York and Chicago
cities_filter = housing_data['City'].isin(['New York', 'Chicago'])
cities_data = housing_data[cities_filter]

# Regression plot for Price vs. Living Space in New York and Chicago
plt.figure(figsize=(12, 6))
sns.lmplot(x='Living Space', y='Price', hue='City', data=cities_data, aspect=2, height=6)
plt.title('Regression Plot: Price vs. Living Space for New York and Chicago')
plt.xlabel('Living Space (sqft)')
plt.ylabel('Price ($)')
plt.show()
```

```python
# Calculating additional metrics for comparison
cities_data['Bed to Bath Ratio'] = cities_data['Beds'] / cities_data['Baths']
cities_comparison_metrics = cities_data.groupby('City').agg({
    'Price Per Living Space': 'mean',
    'Median Household Income': 'median',
    'Address': 'count',  # Number of apartments available
    'Living Space': 'mean',
    'Bed to Bath Ratio': 'mean'
}).reset_index().melt(id_vars='City')

# Plotting the comparisons in a grid
g = sns.FacetGrid(cities_comparison_metrics, col='variable', sharey=False, sharex=False, col_wrap=2, height=4, aspect=1.5)
g.map_dataframe(sns.barplot, x='City', y='value', hue='City', palette='coolwarm')
g.add_legend()
g.set_titles(col_template="{col_name}")
g.set_axis_labels("", "Value")
g.fig.suptitle('Comparison of Metrics Between New York and Chicago', y=1.02)

# Adjusting layout for better readability
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point
import numpy as np

# Load your street map and housing data
street_map = gpd.read_file('/content/geo_export_ff55341a-2adb-403a-9ae3-a80da8a85141.shp')
df = pd.read_csv('/American_Housing_Data_20231209.csv')

# Filter data for specific cities and state
cities = ["New York", "Brooklyn", "Queens Village", "Bronx", "Staten Island"]
state = ["New York"]
filtered_df = df[df['City'].isin(cities) & df['State'].isin(state)]

# Create GeoDataFrame
crs = {"init": "epsg:4326"}
geometry = [Point(xy) for xy in zip(filtered_df["Longitude"], filtered_df["Latitude"])]
geo_df = gpd.GeoDataFrame(filtered_df, crs=crs, geometry=geometry)

# Plotting
fig, ax = plt.subplots(figsize=(15, 15))
street_map.plot(ax=ax, color="grey", alpha=0.3)  # Base map

# Using 'Price' directly for color
scatter = ax.scatter(geo_df['geometry'].x, geo_df['geometry'].y, c=geo_df['Price'], s=50, cmap='inferno', alpha=1)

# Adding colorbar
fig.colorbar(scatter, ax=ax, label='Price')

plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Housing Prices in New York City')
plt.show()
```
